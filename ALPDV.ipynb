```python
import os

# Define paths to your images and labels folders
images_path = '/content/drive/MyDrive/Datasets/LicensePlate/images'
labels_path = '/content/drive/MyDrive/Datasets/LicensePlate/labels'

# List image files (assuming they are .jpg or .png)
image_files = [f for f in os.listdir(images_path) if f.endswith(('.jpg', '.png'))]
# List label files (assuming they are .txt)
label_files = [f for f in os.listdir(labels_path) if f.endswith('.txt')]

# Get the base filenames (without extension) for comparison
image_bases = set(os.path.splitext(f)[0] for f in image_files)
label_bases = set(os.path.splitext(f)[0] for f in label_files)

# Find image files that do not have a corresponding annotation file
missing_annotations = [f for f in image_files if os.path.splitext(f)[0] not in label_bases]

print("Images without corresponding annotation files:")
print(missing_annotations)
```

```python
import os
import random
import shutil

# Define the source folders (clean dataset)
images_src = images_path  # from earlier
labels_src = labels_path  # from earlier

# Define destination base path for the split dataset
split_base = '/content/drive/MyDrive/Datasets/LicensePlateSplit'

# Define subfolders for train, validation, and test for both images and labels
subfolders = {
    'train': {'images': os.path.join(split_base, 'train', 'images'),
              'labels': os.path.join(split_base, 'train', 'labels')},
    'val': {'images': os.path.join(split_base, 'val', 'images'),
            'labels': os.path.join(split_base, 'val', 'labels')},
    'test': {'images': os.path.join(split_base, 'test', 'images'),
             'labels': os.path.join(split_base, 'test', 'labels')}
}

# Function to check if dataset split already exists
def check_existing_split():
    for split in subfolders:
        img_files = os.listdir(subfolders[split]['images'])
        label_files = os.listdir(subfolders[split]['labels'])
        if not img_files or not label_files:  # If any split folder is empty, re-split
            return False
    return True

# If dataset split already exists, skip processing
if check_existing_split():
    print("Dataset split already exists. Skipping processing.")
else:
    print("Splitting dataset...")

    # Create destination directories if they don't exist
    for split in subfolders:
        os.makedirs(subfolders[split]['images'], exist_ok=True)
        os.makedirs(subfolders[split]['labels'], exist_ok=True)

    # Get list of remaining images (after deletion) and shuffle them
    all_images = [f for f in os.listdir(images_src) if f.endswith(('.jpg', '.png'))]
    random.shuffle(all_images)

    # Define split ratios: 70% train, 20% validation, 10% test
    num_images = len(all_images)
    train_count = int(num_images * 0.7)
    val_count = int(num_images * 0.2)
    test_images = all_images[train_count+val_count:]
    train_images = all_images[:train_count]
    val_images = all_images[train_count:train_count+val_count]

    def copy_files(file_list, dest_img_dir, dest_label_dir):
        for file_name in file_list:
            basename = os.path.splitext(file_name)[0]
            src_img = os.path.join(images_src, file_name)
            src_label = os.path.join(labels_src, basename + '.txt')
            dest_img = os.path.join(dest_img_dir, file_name)
            dest_label = os.path.join(dest_label_dir, basename + '.txt')

            shutil.copy(src_img, dest_img)
            if os.path.exists(src_label):
                shutil.copy(src_label, dest_label)
            else:
                print(f"Warning: No label file found for {file_name}")

    # Copy files into the respective split directories
    copy_files(train_images, subfolders['train']['images'], subfolders['train']['labels'])
    copy_files(val_images, subfolders['val']['images'], subfolders['val']['labels'])
    copy_files(test_images, subfolders['test']['images'], subfolders['test']['labels'])

    print("Dataset has been split into Train, Val, and Test sets.")
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the denoised image (or the original image if preferred)
img_path = '/content/drive/MyDrive/Datasets/LicensePlate/test_images/cropped_plate_0.jpg'
image = cv2.imread(img_path)

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply edge detection
edges = cv2.Canny(gray, 50, 150, apertureSize=3)

# Use Hough Line Transform to detect dominant angles
lines = cv2.HoughLinesP(edges, 1, np.pi / 180, threshold=100, minLineLength=100, maxLineGap=10)

angles = []
if lines is not None:
    for line in lines:
        x1, y1, x2, y2 = line[0]
        angle = np.arctan2(y2 - y1, x2 - x1) * 180 / np.pi  # Convert radian to degrees
        angles.append(angle)

# Compute the median angle to avoid outliers
if angles:
    skew_angle = np.median(angles)
else:
    skew_angle = 0  # Default to no rotation if no lines are detected

# Correct the skew by rotating
(h, w) = image.shape[:2]
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, skew_angle, 1.0)
rotated = cv2.warpAffine(image, M, (w, h), flags=cv2.INTER_CUBIC, borderMode=cv2.BORDER_REPLICATE)

# Ensure landscape mode (wider than tall)
if rotated.shape[0] > rotated.shape[1]:  # If the height is greater than width
    rotated = cv2.rotate(rotated, cv2.ROTATE_90_CLOCKWISE)  # Rotate 90 degrees to fix orientation

# Display the final corrected image
cv2_imshow(rotated)
# cv2.waitKey(0)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Convert the skew-corrected image to grayscale
gray = cv2.cvtColor(denoised, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to reduce noise
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Apply Canny edge detection
edges = cv2.Canny(blurred, 50, 150, apertureSize=3)

# Find contours from the edged image
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Initialize variables for license plate
plate_contour = None
max_area = 0
cropped_plate = None

# Filter contours based on aspect ratio and area
for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)
    aspect_ratio = w / float(h)
    area = w * h

    # Condition: Width > Height (landscape mode) and reasonable aspect ratio
    if 2.0 < aspect_ratio < 6.0 and area > 5000:  # Adjust area threshold if needed
        if area > max_area:
            max_area = area
            plate_contour = contour

# If a valid contour is found, crop it
if plate_contour is not None:
    x, y, w, h = cv2.boundingRect(plate_contour)
    cropped_plate = denoised[y:y+h, x:x+w]
else:
    cropped_plate = denoised.copy()  # If no plate found, keep original

# Display the cropped license plate
cv2_imshow(cropped_plate)
# cv2.waitKey(0)
```

```python
import cv2
import pytesseract
from pytesseract import Output
from google.colab.patches import cv2_imshow

# Define DPI (adjust if needed)
DPI = 300
PIXEL_TO_MM = 25.4 / DPI  # Conversion factor

# Convert image to grayscale and binary
gray_enlarged = cv2.cvtColor(enlarged, cv2.COLOR_BGR2GRAY)
_, binary = cv2.threshold(gray_enlarged, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
binary_bgr = cv2.cvtColor(binary, cv2.COLOR_GRAY2BGR)
cv2_imshow(binary_bgr)

# Get image dimensions in pixels and convert to mm
img_height_px, img_width_px = binary.shape
img_width_mm = img_width_px * PIXEL_TO_MM
img_height_mm = img_height_px * PIXEL_TO_MM
print(f"Image Dimensions: Width = {img_width_mm:.2f} mm, Height = {img_height_mm:.2f} mm")

# OCR with detailed character-level data
data = pytesseract.image_to_data(binary, output_type=Output.DICT)

char_dims_mm = []
spaces_mm = []

prev_right_px = None

for i in range(len(data['text'])):
    if data['text'][i].strip() != "":
        x, y, w, h = data['left'][i], data['top'][i], data['width'][i], data['height'][i]
        char_dims_mm.append((w * PIXEL_TO_MM, h * PIXEL_TO_MM))

        # Calculate space between characters in mm
        if prev_right_px is not None:
            space_px = x - prev_right_px
            if space_px > 0:
                spaces_mm.append(space_px * PIXEL_TO_MM)
        prev_right_px = x + w

# Compute average character size in mm
if char_dims_mm:
    avg_char_width_mm = sum(w for w, h in char_dims_mm) / len(char_dims_mm)
    avg_char_height_mm = sum(h for w, h in char_dims_mm) / len(char_dims_mm)
    print(f"Average Character Width: {avg_char_width_mm:.2f} mm")
    print(f"Average Character Height: {avg_char_height_mm:.2f} mm")
else:
    print("No characters detected.")

# Compute average spacing in mm
if spaces_mm:
    avg_space_mm = sum(spaces_mm) / len(spaces_mm)
    print(f"Average Space Between Characters: {avg_space_mm:.2f} mm")
else:
    print("No spaces detected.")
```

```python
import pytesseract
from PIL import Image as PILImage, ImageDraw, ImageFont
import torchvision.transforms as transforms
import torch
import torch.nn as nn
from torchvision import models, datasets
from torch.utils.data import DataLoader
import os
import random
from IPython.display import display
```

```python
arial_font_path = [
    "/content/drive/MyDrive/Font_Check/Arial/ARIALN.TTF",
    "/content/drive/MyDrive/Font_Check/Arial/ARIALNB.TTF",
    "/content/drive/MyDrive/Font_Check/Arial/ARIALNBI.TTF",
    "/content/drive/MyDrive/Font_Check/Arial/ARIALNI.TTF",
    "/content/drive/MyDrive/Font_Check/Arial/arial.ttf",
    "/content/drive/MyDrive/Font_Check/Arial/arialbd.ttf",
    "/content/drive/MyDrive/Font_Check/Arial/arialnbi.ttf",
    "/content/drive/MyDrive/Font_Check/Arial/arialli.ttf",
    "/content/drive/MyDrive/Font_Check/Arial/arialk.ttf"
]

other_fonts = [
    "/content/drive/MyDrive/Font_Check/Not_Arial/couriernew.ttf",
    "/content/drive/MyDrive/Font_Check/Not_Arial/couriernewbold.ttf",
    "/content/drive/MyDrive/Font_Check/Not_Arial/couriernewitalic.ttf",
    "/content/drive/MyDrive/Font_Check/Not_Arial/ocrb.ttf"
]
```

```python
def generate_font_images(output_dir, num_images=100, image_size=(224, 224), text="Sample"):
    os.makedirs(f"{output_dir}/arial", exist_ok=True)
    os.makedirs(f"{output_dir}/not_arial", exist_ok=True)

    for i in range(num_images):
        for label in ["arial", "not_arial"]:
            img = PILImage.new("RGB", image_size, color="white")
            draw = ImageDraw.Draw(img)
            try:
                if label == "arial":
                    font_path = random.choice(arial_font_path)
                else:
                    font_path = random.choice(other_fonts)
                font = ImageFont.truetype(font_path, 32)
                draw.text((10, 90), text, font=font, fill="black")
                img.save(f"{output_dir}/{label}/{label}_{i}.png")
            except:
                continue

output_dir = "/content/drive/MyDrive/Font_Dataset"
generate_font_images(output_dir, num_images=100)

# Training
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])
dataset = datasets.ImageFolder(output_dir, transform=transform)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

model = models.resnet18(pretrained=True)
model.fc = nn.Linear(model.fc.in_features, 2)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

for epoch in range(5):
    model.train()
    running_loss = 0
    for imgs, labels in dataloader:
        imgs, labels = imgs.to(device), labels.to(device)
        outputs = model(imgs)
        loss = criterion(outputs, labels)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    print(f"Epoch {epoch+1}, Loss: {running_loss/len(dataloader):.4f}")

# Save model
torch.save(model.state_dict(), "/content/drive/MyDrive/font_detector.pth")
```

```python
from PIL import Image, ImageDraw, ImageFont
import numpy as np
import torch
import random

def ocr_and_predict_font(image=binary_bgr, check_as="arial"):
    # Convert OpenCV image (NumPy array) to PIL Image
    if isinstance(image, np.ndarray):
        image = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

    original_image = image.convert("RGB")
    extracted_text = pytesseract.image_to_string(original_image).strip()

    if not extracted_text:
        print("❌ No text detected in image.")
        return "No text found"

    print(f"📝 Extracted Text: {extracted_text}")

    test_img = Image.new("RGB", (224, 224), "white")
    draw = ImageDraw.Draw(test_img)

    font_path = random.choice(arial_font_path if check_as == "arial" else other_fonts)
    try:
        font = ImageFont.truetype(font_path, 32)
    except:
        print(f"❌ Font not found or failed to load: {font_path}")
        return "Font rendering failed"

    draw.text((10, 90), extracted_text, font=font, fill="black")

    img_tensor = transform(test_img).unsqueeze(0)
    with torch.no_grad():
        output = model(img_tensor)
        pred = torch.argmax(output, dim=1).item()

    return "Arial" if pred == 0 else "Not Arial"
```

```python
# Directly use the image instance (e.g., binary_bgr)
print("🔍 Running OCR and font prediction on in-memory image...")

result = ocr_and_predict_font(image=binary_bgr, check_as="arial")  # or "not_arial"
print(f"🧠 Prediction Result: {result}")

# Optionally display the image
from PIL import Image as PILImage
display(PILImage.fromarray(cv2.cvtColor(binary_bgr, cv2.COLOR_BGR2RGB)))
```
